```python
#juntar grafos
G0 = nx.Graph([(0, 1), (1, 2), (2, 0)])
G1 = nx.Graph([(0, 1), (1, 2), (2, 3), (3, 4), (2, 4)])
#uniao
G2 = nx.disjoint_union(G0, G1) #vai considerar que 1 != 1
#atribui lable difernte
pos = nx.spring_layout(G2)
nx.draw(G2, with_labels=True, node_size=500, font_size=10, pos=pos)
plt.show(True)
```

```python
G3 = nx.compose(G1, G2) #assume que 1 em G1 = 1 em G2
#uniao juntando quem tem mesmo lable
pos = nx.spring_layout(G3)
nx.draw(G3, with_labels=True, node_size=500, font_size=10, pos=pos)
plt.show(True)
```

```python
G = G3.to_undirected() #transformar pra sem direcao
pos = nx.spring_layout(G)
nx.draw(G, with_labels=True, node_size=500, font_size=10, pos=pos)
plt.show(True)
G = G.to_directed()#transformar pra com direcao
pos = nx.spring_layout(G)
nx.draw(G, with_labels=True, node_size=500, font_size=10, pos=pos)
plt.show(True)
```

Gerar dados que seguem uma lei de potência pela simulação de monte carlo e método da transformada inversa

```python
import numpy as np
import matplotlib.pyplot as plt

alpha = 3
xmin = 1
n = 10000
data = []

for i in range(0, n):
  u = np.random.uniform(0, 1)
  x = xmin*(1-u)**(-1/(alpha-1))
  data.append(x)
```

Distribuição Lei de Potência (estimar alfa)

```python
import powerlaw #permite estimar xmin (distancia de ...) e alfa
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(101)
```

```python
n = len(data)

results = powerlaw.Fit(data)
xmin = results.power_law.xmin
alpha_hat = results.power_law.alpha
print('alpha estimado:', alpha_hat)
print('xmin estimado:', xmin)

fig = plt.figure(figsize=(8, 6))
nb = 500
p, bins = np.histogram(data, bins=nb, density=True)
bins_mean = [0.5*(bins[i]+bins[i+1])
            for i in range(len(p))]
#curva teorica
y = np.linspace(1, np.max(data), 1000)
f = ((alpha-1)/xmin)*(y/xmin)**(-alpha)
plt.plot(y, f, '--', color='black')
plt.xlabel('x', fontsize=20)
plt.ylabel('f(x)', fontsize=20)
plt.show(True)

#log-log
plt.loglog(bins_mean, p, 'o', color = 'gray', alpha=0.6, linewidth=1)

y = np.linspace(xmin, np.max(data), 100)
f_hat = (alpha_hat-1)*xmin**(alpha_hat-1)*y**(-alpha_hat)
plt.plot(y, f_hat, '--', color='gray', label='Modelo estimado')
plt.xlabel('x', fontsize=20)
plt.ylabel('f(x)', fontsize=20)
plt.legend(fontsize=15)
plt.show(True)
```

```python
#distribuição acumulada complementar para reduzir as discrepâncias - se aproxima mais da distribuição da lei de potencia
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
count, bins_count = np.histogram(data, bins=100)
pdf = count/sum(count)
edf = 1-np.cumsum(pdf)
plt.loglog(bins_count[1:-1], edf[0:-1], 'o', label='CDF', color='gray', alpha=0.6, linewidth=1)
y = np.linspace(1, np.max(data), 1000)
f = ((alpha-1)/xmin)*(y/xmin)**(-alpha)
plt.plot(y, f, '--', color='black')
plt.xlabel('x', fontsize=20)
plt.ylabel('f(x)', fontsize=20)
plt.show(True)
```

```python
#estimação de verossimilhança
#alfa^ = 1 + n/sum(log(xi/xmin))
s = 0
for i in range(0, n):
  s += np.log(data[i]/xmin)
alpha_hat = i+n/s
print(alpha)
print(alpha_hat)
```
